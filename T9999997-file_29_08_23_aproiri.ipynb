```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df
```

```python
data = df.groupby(['Transaction','Item'])['Item'].count().reset_index(name='count')
data
```

```python
#pivot table
data_final = data.pivot_table(index='Transaction',columns='Item',values='count').fillna(0.0)
data_final['scandinavian']
```

```python
def change(s):
    if s<=0.0:
        return 0
    elif s>=1.0:
        return 1
df_new = data_final.applymap(change)
df_new
```

```python
frequent_items = apriori(df_new,min_support=0.05,use_colnames=True)
frequent_items
```

```python
association_rules(frequent_items,metric='confidence',min_threshold=0.01)
```
